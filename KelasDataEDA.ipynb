```python
import pandas as pd
import numpy as np
import seaborn as sns ##Visualiasi
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/KennethParulian/PythonEDAEnigmacamp/main/dummydata_kendaraan.csv'
df = pd.read_csv(url)
```

```python
stok12 = df['cust_first_name'].groupby(df['brand']).count().nlargest(5)
plt.barh(stok12.index, stok12 , color =['#FF0000'], alpha = 0.10)
plt.ylabel('nama produk')
plt.xlabel('jumlah')
plt.title('data terjual')
plt.savefig('kelasEDA.jpg')
plt.show()
```

```python
sns.countplot(x='model', data=df , order = pd.value_counts(df['produce_year']).iloc[:10].index)
```

```python
with sns.axes_style(style='dark'):
  g = sns.catplot('brand', 'model', 'price', data=df, kind='box')
  g.set_axis_labels('brand', 'model')
```
